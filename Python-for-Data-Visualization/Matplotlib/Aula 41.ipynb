# MatPlotLib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

# Functional Method

```python
plt.plot(x,y, 'r-')
plt.ylabel('Y Label')
plt.xlabel('X Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')

plt.show()
```

# OO - Orientado a objetos

```python
fig = plt.figure()

axes = fig.add_axes([0.3,0.3,1,1])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)

axes2 = fig.add_axes([.1,.4,.4,.4])
axes2.plot(y,x)
```
